```python
import textwrap
def wrap_text(text, width=90): #preserve_newlines
    # Split the input text into lines based on newline characters
    lines = text.split('\n')

    # Wrap each line individually
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

    # Join the wrapped lines back together using newline characters
    wrapped_text = '\n'.join(wrapped_lines)

    return wrapped_text
```

```python
def fetch_html_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Remove unwanted elements like scripts and styles
    for tag in soup(["script", "style"]):
        tag.decompose()

    # Extract and retain structure
    html_content = ""
    for tag in soup.find_all(["h1", "h2", "h3", "p", "ul", "ol", "li", "b", "i"]):
        if tag.name == "h1":
            html_content += f"<h1>{tag.text.strip()}</h1>\n"
        elif tag.name == "h2":
            html_content += f"<h2>{tag.text.strip()}</h2>\n"
        elif tag.name == "h3":
            html_content += f"<h3>{tag.text.strip()}</h3>\n"
        elif tag.name == "p":
            html_content += f"<p>{tag.text.strip()}</p>\n"
        elif tag.name == "ul":
            html_content += "<ul>\n"
            for li in tag.find_all("li"):
                html_content += f"<li>{li.text.strip()}</li>\n"
            html_content += "</ul>\n"
        elif tag.name == "ol":
            html_content += "<ol>\n"
            for li in tag.find_all("li"):
                html_content += f"<li>{li.text.strip()}</li>\n"
            html_content += "</ol>\n"
        elif tag.name == "b":
            html_content += f"<b>{tag.text.strip()}</b>\n"
        elif tag.name == "i":
            html_content += f"<i>{tag.text.strip()}</i>\n"

    return html_content
```

```python
def detect_drive_links(text):
    drive_link_pattern = r"https://drive\.google\.com/file/d/([a-zA-Z0-9_-]+)/view"
    return re.findall(drive_link_pattern, text)
```

```python
def fetch_text_from_drive(service, file_id):
    try:
        # Get file metadata
        request = service.files().get(fileId=file_id, fields="name, mimeType")
        file_metadata = request.execute()

        file_name = file_metadata["name"]
        mime_type = file_metadata["mimeType"]

        # Handle file content download
        request = service.files().get_media(fileId=file_id)
        file_data = io.BytesIO()
        downloader = MediaIoBaseDownload(file_data, request)
        done = False
        while not done:
            _, done = downloader.next_chunk()
        file_data.seek(0)

        # Extract text based on file type
        if mime_type == "application/pdf":
            from PyPDF2 import PdfReader
            reader = PdfReader(file_data)
            return " ".join(page.extract_text() for page in reader.pages)
        elif mime_type.startswith("text/"):
            return file_data.read().decode("utf-8")
        elif mime_type in ["application/vnd.openxmlformats-officedocument.wordprocessingml.document"]:
            from docx import Document
            doc = Document(file_data)
            return " ".join(paragraph.text for paragraph in doc.paragraphs)
        else:
            return f"Unsupported file type: {mime_type}"
    except Exception as e:
        print(f"Error fetching file {file_id}: {e}")
        return ""
```

```python
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain.prompts import ChatMessagePromptTemplate, PromptTemplate
```

```python
prompt = ChatPromptTemplate(input_variables=['original_query'],
                            messages=[SystemMessagePromptTemplate(prompt=PromptTemplate(input_variables=[],template='You are a helpful assistant that generates multiple search queries based on a single input query.')),
                            HumanMessagePromptTemplate(prompt=PromptTemplate(input_variables=['original_query'], template='Generate multiple search queries related to: {question} \n OUTPUT (4 queries):'))])
```

```python
original_query = "Tell me the what is question 2 in the wireshark-UDP-v8.0 pdf"
```

```python
generate_queries = (
    prompt | llm | StrOutputParser() | (lambda x: x.split("\n"))
)
```

```python
generate_queries
```

```python
from langchain.load import dumps, loads


def reciprocal_rank_fusion(results: list[list], k=60):
    fused_scores = {}
    for docs in results:
        # Assumes the docs are returned in sorted order of relevance
        for rank, doc in enumerate(docs):
            doc_str = dumps(doc)
            if doc_str not in fused_scores:
                fused_scores[doc_str] = 0
            previous_score = fused_scores[doc_str]
            fused_scores[doc_str] += 1 / (rank + k)

    reranked_results = [
        (loads(doc), score)
        for doc, score in sorted(fused_scores.items(), key=lambda x: x[1], reverse=True)
    ]
    return reranked_results
```

```python
ragfusion_chain = generate_queries | retriever.map() | reciprocal_rank_fusion
```

```python
def calculate_bleu(reference_text, qa_pairs):
    references = [[answer.split()] for answer in reference_text]
    answers = [answer for _, answer in qa_pairs]  # List of lists for multiple references
    candidates = [answer.split() for answer in answers]  # Tokenized candidate responses

    # Calculate BLEU score
    bleu_score = corpus_bleu(references, candidates)
    return bleu_score
```

```python
def calculate_rouge(true_answers, generated_answers):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    rouge1_scores = []
    rouge2_scores = []
    rougeL_scores = []

    for reference, candidate in zip(true_answers, generated_answers):
        scores = scorer.score(candidate, reference)  # Compare LLM/RAG-Fusion output with ground truth
        rouge1_scores.append(scores['rouge1'].fmeasure)
        rouge2_scores.append(scores['rouge2'].fmeasure)
        rougeL_scores.append(scores['rougeL'].fmeasure)

    avg_rouge1 = sum(rouge1_scores) / len(rouge1_scores)
    avg_rouge2 = sum(rouge2_scores) / len(rouge2_scores)
    avg_rougeL = sum(rougeL_scores) / len(rougeL_scores)

    return avg_rouge1, avg_rouge2, avg_rougeL
```
